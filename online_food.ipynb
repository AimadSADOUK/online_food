```python
# Importing the library matplotlib
import matplotlib.pyplot as plt

# Counting the occurrences of each gender
G = df['Gender'].value_counts()

# Extracting values and labels from the value counts
A = G.values  # Counts of each gender
B = G.index   # Names of each gender

# Define colors for the pie chart
colors = ['skyblue', 'pink']

# Define the extent to which a wedge is separated from the center of the pie
explode = (0.1, 0)

# Create a new figure for the plot
plt.figure(figsize=(6, 6))

# Plotting the pie chart
plt.pie(A, labels=B, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True)

# Adding a title to the pie chart
plt.title('Number total of genders', fontsize='xx-large')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Counting the occurrences of each marital status category
MS_counts = df['Marital Status'].value_counts()

# Extracting the counts and category names
MS_counts_values = MS_counts.values  # Counts of each marital status category
MS_counts_index = MS_counts.index    # Names of each marital status category

# Define colors for the donut plot
colors = ['skyblue', 'lightgray', 'pink']

# Define the extent to which a wedge is separated from the center of the donut plot
explode = (0, 0, 0)

# Create a new figure for the plot
plt.figure(figsize=(6, ))

# Plotting the donut plot
plt.pie(MS_counts_values, labels=MS_counts_index, colors=colors, explode=explode,
        shadow=True, autopct='%1.1f%%', wedgeprops=dict(width=0.3))

# Adding a circle at the center to create the donut shape
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that the donut plot is drawn as a circle
plt.axis('equal')

# Adding a title
plt.title('Marital Status Distribution')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is defined and contains the necessary data

# Calculate value counts for 'Occupation'
occupation_counts = df['Occupation'].value_counts()

# Extracting index and values from the value counts
itemName = occupation_counts.index  # Names of each occupation
itemValue = occupation_counts.values  # Counts of each occupation

colors = ['LightSalmon', 'PeachPuff', 'NavajoWhite', 'SandyBrown']

# Create a new figure for the plot
plt.figure(figsize=(10, 6))

# Plotting the bar chart
plt.barh(itemName, itemValue, color=colors)

# Adding labels and title to the plot
plt.xlabel('Count', fontsize='x-large')
plt.ylabel('Occupation', fontsize='x-large')
plt.title('The count of each Occupation', fontsize='xx-large')

# Show the values over the columns
for i in range(len(itemName)):
    plt.text(itemValue[i] + 0.5, i, itemValue[i], ha='left', va='center')

plt.xticks(rotation=45)

# Display the plot
plt.gca().invert_yaxis()  # Invert the y-axis to reverse the order of bars
plt.show()
```
